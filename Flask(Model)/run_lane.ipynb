```python
### 차선 분류 데이터 추출
# !pip install ujson
import ujson
import os
import sys
# if "/opt/ros/kinetic/lib/python2.7/dist-packages" in sys.path:
#     sys.path.remove("/opt/ros/kinetic/lib/python2.7/dist-packages")
import cv2
import numpy as np
import shutil
from shutil import rmtree  

target_img_dir = "./dataset/images"
target_lb_dir = "./dataset/labels"



# if (os.path.isdir(target_img_dir) == True):
#   shutil.rmtree(target_img_dir)
# else:
#   os.mkdir(target_img_dir)

# if (os.path.isdir(target_lb_dir) == True):
#   shutil.rmtree(target_lb_dir)
# else:
#   os.mkdir(target_lb_dir)


laneType = [{"white","blue"},{"dotted","solid"}]
mapping_dict = {
    "white-dotted":3,
    "white-solid":4,
    # "yellow-solid":5
    # "yellow-dotted":5,--> 예외처리
    #"blue-dotted":7, # "yellow-dotted":5,--> 예외처리
    #"blue-solid":8--> 예외처리
    }

# lightstates = ["red","yellow","green"]
# mapping_dict = {
#     "100":"0",
#     "010":"1",
#     "001":"2",
# }

visit = dict()

# 원본파일 가져올 경로
path = "C:/Users/smhrd/Desktop/dataset/lane/image"

imgnames = os.listdir(path)

for imgname in imgnames:
    imgpath = path+'/'+imgname
    jsonpath=imgpath.replace("/image","/json").replace(".jpg",".json").replace(".png",".json")
   
    img = None
    with open(jsonpath,"r") as json:
      json_data = ujson.load(json)
    
    index = 0

    imsize_h, imsize_w=json_data["image"]["image_size"]
    imgname = json_data["image"]["file_name"]

    new_imgpath = target_img_dir+"/"+ imgname
    new_lbpath = new_imgpath.replace("/images","/labels").replace(".jpg", ".txt")
   

    for d in json_data["annotations"]:
      if d["class"] == "traffic_lane" :
            
        # laneType = [{"white","yellow","blue"},{"dotted","solid"}]
        #                  0,    1,       2        0,    1
        lane1 = d['attributes'][0]['value']
        lane2 = d['attributes'][1]['value']

        # if light=="0000" or light=="0001"or light=="1010"or light=="1100":
        # continue
        lane = lane1+"-"+lane2
        
        if lane=="white-dotted" or lane=="white-solid": #or lane=="yellow-solid" :
            lane = mapping_dict[lane]        
            print(imgname)  
            with open(new_lbpath, 'a') as f:
                x0= d["data"][0]["x"]
                y0= d["data"][0]["y"]
                x1= d["data"][1]["x"]
                y1= d["data"][1]["y"]
                xywh = [((x1+x0)/2)/imsize_w, (((y1+y0)/2))/imsize_h, abs(x1-x0)/imsize_w, abs(y1-y0)/imsize_h]  # xywh normalized
                line = int(lane),*xywh  # YOLO format (class_index, xywh)
                f.write(('%g ' * len(line)).rstrip() % line + '\n')
            print(line)
       
        shutil.copy2(imgpath,new_imgpath)        
    
    index += 1
```
